### Hypothesis testing 


```python
contigency= pd.crosstab(data['Tags'], data['What is your current occupation']) 
contigency
```

```python
from scipy.stats import chi2_contingency
```

```python
# Chi-square test of independence. 
c, p, dof, expected = chi2_contingency(contigency)
# Print the p-value
print(p,c,dof)
```

```python
import scipy.stats

#find Chi-Square critical value
scipy.stats.chi2.ppf(1-.05, df=125)
```

we reject the null Hypothesis and say there is a relationship between variables also chic square sts> critical value

```python
from scipy import stats

t_value,p_value=stats.ttest_ind(data['Total Time Spent on Website'],data['Converted'])
```

```python
print('Test statistic is %f'%float("{:.6f}".format(t_value)))

print('p-value for two tailed test is %f'%p_value)

alpha = 0.05

if p_value<=alpha:

    print('Conclusion','n','Since p-value(=%f)'%p_value,'<','alpha(=%.2f)'%alpha,'we reject the null hypothesis')

else:

    print('Conclusion','n','Since p-value(=%f)'%p_value,'>','alpha(=%.2f)'%alpha)
```

we reject the null Hypothesis and say there is a relationship between variables also chic square sts> critical value
